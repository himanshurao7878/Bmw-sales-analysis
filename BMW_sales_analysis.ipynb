 Insight:

```python
models = df1.groupby('Model')
len( df1.groupby('Model'))
```

```python
models['Sales_Volume'].describe()
```

```python
models['Sales_Volume'].sum().sort_values(ascending = False).head(1)
```

```python
explode = [0.1 if model == '7 Series' else 0 for model in models['Sales_Volume'].sum().sort_values(ascending = False).index]
models['Sales_Volume'].sum().sort_values(ascending = False).plot(kind = 'pie',autopct='%0.1f%%',explode = explode )

plt.title('Sales Volume Distribution by BMW Model')
```

Question

what is the average price of each bmw model and which one is the most expensive on average

ans:  7 series is the most expensive on average


```python
models = df1.groupby('Model')
```

```python
avg_price = models['Price_USD'].mean()
print(avg_price)
```

```python
Model = avg_price.idxmax()
Price = avg_price.max()
print(Model,Price)
```

```python
models['Price_USD'].sum().sort_values(ascending = False).plot(kind = 'bar')
```

```python
avg_price.sort_values().plot(kind='bar')
```

Question

```python
Region = df1.groupby('Region')
Name = Region['Sales_Volume'].sum().idxmax()
Value = Region['Sales_Volume'].sum().max()
print(Name ,Value)
```
